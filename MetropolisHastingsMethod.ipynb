```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import random
```

```python
## Metropolis Hastings Algorithm 


def p(x):
    pr=0.3*np.exp(-0.2*(x**2))+0.7*np.exp(-0.2*((x-10)**2))
    return pr
    
######## density plot of p(x) ########

grid_x=np.arange(-10,20,0.01)

ax = plt.gca()
ax.axes.yaxis.set_visible(False)
plt.title('Density plot of p(x)')
plt.plot(grid_x,p(grid_x),'r')
######## Simulation 1 - when sigma=1 ########

T=10000 # Total iteration number
sigma=20
MH_x_1=np.random.normal(size=1,loc=0,scale=20) # Set initial value

for i in range(1,T+1):
    # Draw candidate
    candi_x = np.random.normal(size=1,loc=MH_x_1[i-1],scale=sigma)
    
    # Compute acceptance rate
    nume = p(candi_x)/st.norm(MH_x_1[i-1], sigma).pdf(candi_x)
    deno = p(MH_x_1[i-1])/st.norm(candi_x,sigma).pdf(MH_x_1[i-1]) 
    alpha = min(nume/deno,1)
    
    #Compare acceptance rate with U
    U = np.random.uniform(size=1)
    if U<=alpha: 
        MH_x_1=np.append(MH_x_1,candi_x)
    else:
        MH_x_1=np.append(MH_x_1,MH_x_1[i-1])

MH_x_1=np.delete(MH_x_1,0) # Drop initial value

######## Histogram of samples ########

grid_x=np.arange(-10,20,0.01)
a= sum(p(grid_x))
print(a)
plt.plot(grid_x,p(grid_x)/sum(p(grid_x))*100, 'b')
plt.hist(MH_x_1[T//2:],bins=40,histtype='step', density = True)
ax = plt.gca()
ax.axes.yaxis.set_visible(False)
plt.title(r'Histogram of samples when $\sigma=1$')
```

# 모집단 분포 추정

```python



np.random.seed(1)
a=2.7
b=6.3
naccept=0
niters=1000
samples=np.zeros(niters+1)
samples[0]=st.uniform.rvs()

for i in range(niters):
  xt=samples[i]
#   y=st.gamma.rvs(a=xt,scale=1)
  y=st.uniform.rvs()
  #y = np.random.normal(size=1,loc=xt,scale=5)
  rho = min(1, st.beta(a,b).pdf(y)/st.beta(a,b).pdf(xt))
  u=st.uniform.rvs()
  if u <rho:
    naccept+=1
    samples[i+1] = y
  else:
    samples[i+1] =xt
nmcmc = len(samples)//2

print ("Efficiency=", naccept/niters)



post = st.beta(a,b)
x=np.linspace(0,1,200)
plt.figure (figsize=(12,9))
plt.hist(samples[nmcmc:],40, histtype='step', density=True, linewidth=1, label = "Distribution of posterior samples")
plt.plot(x, post.pdf(x), c='red',linestyle='--',alpha=0.5,label='True posterior')
plt.xlim([0,1]);
plt.legend(loc='best');
```

# MCMC 로 theta0 분포 approximating (in discrete)

```python
def Pi1(Xa,Xb,M, K1,K2,K3,K0):
    
    C1 = K2/K3
    C2 = K3/K1
    # solve the equations 
    
    numerator = M*(C1**Xa)*(C2**Xb)
    denominator1 = 1
    denominator2 = 1
    for i in range(int(Xa)):
        denominator1 *= ((K1*(i+1))/(K1+K0*(i+1)))
    for j in range(int(Xb)):
        denominator2 *= (j+1)
    denominator = ((K1+K0*Xa)/K1)*denominator1*denominator2
        
    return numerator/denominator
    
def likelihood(y,k0):
    result =1
    for i in range(len(y)):
        result *=  Pi1(y[i][0], y[i][1],M,k1,k2,k3, k0)
    return result


#np.random.seed(1)

naccept=0
niters=10000
# initial condition
M = 1
k1 = 1
k2 =1 
k3 = 1


a = [0,0.05,0.10,0.15,0.20,0.25,0.30,0.35,0.40,0.45,0.50,0.55,0.60,0.65,0.70,0.75,0.80,0.85,0.90,0.95]
xk = np.array(a)
#a = 10*a
#xk = (0.1,0.4,0.5,0.7)
#yk = (1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9,1/9)
yk = (1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20,1/20)
# yk is prior 

def f(x):
    index = a.index(x)
    return yk[index]



#np.random.choice(elements, 10, p=probabilities)

samples=np.zeros(niters+1)
samples[0]=0.10
for i in range(niters):
    #custm = st.rv_discrete(name='custm', values=(xk, yk))
    k0_0=samples[i]
      # y=st.gamma.rvs(a=xt,scale=1)
    t=st.uniform.rvs()
    counting =1
    while True:
        if t <= counting/20:
            k0_1 = 0.05*(counting-1)
            break
        else:
            counting += 1
        
  #y = np.random.normal(size=1,loc=xt,scale=5)
  #print(custm.pmf(y))

    rho = min(1, (likelihood(y,k0_1))/(likelihood(y,k0_0)))
    u = st.uniform.rvs()
    if u <rho:
        naccept+=1
        samples[i+1] = k0_1
    else:
        samples[i+1] =k0_0
    #print(samples[i+1])
        
        
        
nmcmc = len(samples)//2

print ("Efficiency=", naccept/niters)

plt.figure (figsize=(12,9))
plt.hist(samples[nmcmc:],10, histtype='step', density=True, linewidth=1, label = "Distribution of samples")
#plt.plot(xk, like(xk)/sum(like(xk))*100, c='red',linestyle='--',alpha=0.5,label='True posterior')
plt.xlim([0,1]);
plt.legend(loc='best');    
```
